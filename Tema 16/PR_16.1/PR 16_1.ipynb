__Importamos las librerias__

```python
from urllib.request import urlopen
from urllib.error import HTTPError
from bs4 import BeautifulSoup

import json
```

__Definimos la url de la web__

```python
url = r"https://quotes.toscrape.com/"
```

# 1. Apartado A

_Extrae y almacena en un fichero JSON que llamaremos “citas_A.josn” los siguientes datos: Id (número secuencial que generará el programa), autor (nombre del autor), cita (texto de la cita), etiquetas (array con las tags), tamaño (longitud del texto de la etiqueta)_

* Obtenemos el html en bruto

```python
try:
    html = urlopen(url)
except HTTPError as e:
    print("Página no encontrada")
```

* La convertimos en un objeto BeautifulSoup

```python
bs = BeautifulSoup(html.read(), 'html.parser')
```

* Obtenemos la etiqueta padre

```python
try:
    padre_quotes = bs.find_all('div', attrs={'class': 'row'})[1].find('div', attrs={'class': 'col-md-8'})
except AttributeError as e:
    print("Etiqueta no encontrada")
```

* De ella sacamos las quotes

```python
try:
    quotes = padre_quotes.find_all('div', attrs={'class': ['row', 'quote']})
except AttributeError as e:
    print("Etiqueta no encontrada")
```

* Extraemos los datos requeridos

```python
quotes_data = []
id_cont = 1
for quote in quotes:
    # Reseteamos los datos
    quote_data = {"id": 0, "autor": "", "cita": "", "etiquetas": [], "tamano": 0}

    # Id
    quote_data["id"] = id_cont
    id_cont += 1

    # Autor
    autor = quote.find('small', attrs={'class': 'author'}).string
    quote_data["autor"] = autor

    # Cita
    cita = quote.find('span', attrs={'class': 'text'}).string
    quote_data["cita"] = cita

    # Etiquetas
    etiquetas_html = quote.find('div', attrs={'class': 'tags'}).find_all('a', attrs={'class': 'tag'})
    etiquetas = []
    for etiqueta in etiquetas_html:
        etiquetas.append(etiqueta.string)
    quote_data["etiquetas"] = etiquetas

    # Tamaño
    quote_data["tamano"] = len(cita)

    # Lo insertamos 
    quotes_data.append(quote_data)
```

* Lo escribimos en el json

```python
with open(r"./json/citas_A.json","w") as archivo:
    json.dump(quotes_data, archivo, indent=4)
```

# 2. Apartado B

_Partiendo del ejercicio anterior, para cada cita, siguiendo y analizando el enlace que nos lleva a la página del autor (about), añade al JSON del ejercicio anterior los siguientes campos: Fecha ( fecha de nacimiento del autor), localidad (localidad de nacimiento del autor) y url (dirección del enlace de su página). Almacena los datos en un fichero JSON que llamaremos “citas_B.josn”_

* Obtenemos el html en bruto

```python
try:
    html = urlopen(url)
except HTTPError as e:
    print("Página no encontrada")
```

* La convertimos en un objeto BeautifulSoup

```python
bs = BeautifulSoup(html.read(), 'html.parser')
```

* Obtenemos la etiqueta padre

```python
try:
    padre_quotes = bs.find_all('div', attrs={'class': 'row'})[1].find('div', attrs={'class': 'col-md-8'})
except AttributeError as e:
    print("Etiqueta no encontrada")
```

* De ella sacamos las quotes

```python
try:
    quotes = padre_quotes.find_all('div', attrs={'class': ['row', 'quote']})
except AttributeError as e:
    print("Etiqueta no encontrada")
```

* Extraemos los datos requeridos. Insertando también los datos del autor.

```python
quotes_data = []
id_cont = 1
for quote in quotes:
    # Reseteamos los datos
    quote_data = {"id": 0, "autor": "", "cita": "", "etiquetas": [], "tamano": 0}

    # Id
    quote_data["id"] = id_cont
    id_cont += 1

    # Autor
    autor = quote.find('small', attrs={'class': 'author'}).string
    quote_data["autor"] = autor

    # Cita
    cita = quote.find('span', attrs={'class': 'text'}).string
    quote_data["cita"] = cita

    # Etiquetas
    etiquetas_html = quote.find('div', attrs={'class': 'tags'}).find_all('a', attrs={'class': 'tag'})
    etiquetas = []
    for etiqueta in etiquetas_html:
        etiquetas.append(etiqueta.string)
    quote_data["etiquetas"] = etiquetas

    # Tamaño
    quote_data["tamano"] = len(cita)

    # Obtenemos la información avanzada del autor
    autorUrl = quote.a.attrs['href']

    try:
        htmlAutor = urlopen(f"{url}{autorUrl}")
        bsAutor = BeautifulSoup(htmlAutor.read(), 'html.parser')

        quote_data["fechaAutor"] = bsAutor.find('span', {'class': "author-born-date"}).string
        quote_data["localidadAutor"] = bsAutor.find('span', {'class': "author-born-location"}).string.removeprefix("in ")

    except HTTPError as e:
        print(f"Detalles de {autor} no encontrados")


    # Lo insertamos 
    quotes_data.append(quote_data)
```

* Lo escribimos en el json

```python
with open(r"./json/citas_B.json","w") as archivo:
    json.dump(quotes_data, archivo, indent=4)
```
