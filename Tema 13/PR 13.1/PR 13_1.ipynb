__Importamos las librerias__

```python
import requests
import json
import pandas as pd
```

__Definimos la url de la API__

```python
url_busqueda = "https://www.thecocktaildb.com/api/json/v1/1/search.php"
url_filtro = "https://www.thecocktaildb.com/api/json/v1/1/filter.php"
```

# 1. Apartado A

_Realiza un programa en Python que a partir de los datos devueltos por la consulta nos muestre por pantalla los nombres de las categorías y el total de las mismas._

* Llamamos a la api

```python
respuesta = requests.get(url_busqueda, params={"s": "martini"})
```

* Extraemos la consulta

```python
consulta = json.loads(respuesta.content)["drinks"]
```

* Generamos el resultado esperado

```python
categorias = {}
for cocktail in consulta:
    if cocktail["strCategory"] in categorias:
        categorias[cocktail["strCategory"]] += 1
    else:
        categorias[cocktail["strCategory"]] = 1
```

* Mostramos las categorias

```python
for categoria, cantidad in categorias.items():
    print(f"{categoria}: {cantidad}")
```

# 2. Apartado B

_1. Añade al programa del ejercicio anterior una nueva funcionalidad: por cada una de las categorías genera un archivo JSON con el nombre de la misma (p.e.: ”shake.json”)que contenga todos sus cócteles –investiga en la documentación que endpoint has de utilizar-._

* Recorremos las categorias

```python
for categoria, cantidad in categorias.items():
    categoria_format = categoria.replace(" ", "_")
    
    cocktails = json.loads(
            requests.get(url_filtro, params={"c": categoria_format}).content
        )["drinks"]
    
    with open(f"cockteles/{categoria_format}.json", 'w') as archivo_nuevo:
        json.dump(cocktails, archivo_nuevo)
```

_3. Realiza la misma operación que en el ejercicio B.1, pero en esta ocasión almacena todos los datos de los cócteles en un único archivo con formato csv que contendrá para cada cóctel cuatro campos: su Id, nombre, imagen y categoría. La primera fila del archivo ha de incorporar el nombre de los campos._

```python
idCoctel = []
nombreCoctel = []
imagenCoctel = []
categoriasCoctel = []

for categoria, cantidad in categorias.items():
    categoria_format = categoria.replace(" ", "_")
    
    cocktails = json.loads(
            requests.get(url_filtro, params={"c": categoria_format}).content
        )["drinks"]
    
    for i in cocktails:
        llamada = requests.get(url_busqueda, params={"s": i["strDrink"]})
        if llamada.ok:
            coctel = json.loads(
                llamada.content
            )["drinks"]
        
        if coctel is not None:
            coctel = coctel[0]
            idCoctel.append(coctel["idDrink"]),
            nombreCoctel.append(coctel["strDrink"]),
            imagenCoctel.append(coctel["strDrinkThumb"]),
            categoriasCoctel.append(coctel["strCategory"])

cocktails_csv = {"id": idCoctel, "nombre": nombreCoctel, "imagen": imagenCoctel, "categoria": categoriasCoctel}

df = pd.DataFrame(cocktails_csv)

df.to_csv('cocktails.csv', index=True, encoding='utf-8')
```
